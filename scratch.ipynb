```python
import bingo
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tot_ppl = [1] + list(range(5, 101, 5))
picked = range(5, 16)
```

```python
ps = [
    (n, p, bingo.p_anyone_normal_bingo(n, p))
    for n in picked
    for p in tot_ppl
]
```

```python
plt.figure(figsize=(20,10))
for p in tot_ppl:
    tmp = [x for x in ps if x[1] == p]
    plt.plot([x[0] for x in tmp], [y[2] for y in tmp], label=p)
plt.legend()
plt.title('Probability of someone getting bingo on the nth pick for various number of players')
```

```python
plt.figure(figsize=(20,10))
for n in picked:
    tmp = [x for x in ps if x[0] == n]
    plt.plot([x[1] for x in tmp], [y[2] for y in tmp], label=n)
plt.legend()
plt.title('Probability of someone getting bingo given x players on various number picks')
```
